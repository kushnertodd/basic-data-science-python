# Types of Data
[Data](https://www.ibm.com/think/topics/data) are facts, figures, observations, symbols, and other useful information. The purpose of data analysis is to make conclusions from data that can be used in decision making. There are two major types of data: qualitative and quantitative. Qualitative data are descriptive, non-numerical characteristics or concepts that have discrete or a limited set of values. Quantitative data are measurements that generally have continuous and unlimited values. Each of these are divided into subtypes.

## Qualitative Data
Qualitative data has two classes:  categorical or nominal, and ordinal.
- Categorical data consists of a finite number of labels or names that are separate and have no relation to each other, like colors: “red”, “green”, “yellow”, or states: “New York”, “Missouri”, “Arizona”.
- Ordinal data values have a natural ordering or ranking, like days of the week: “Monday”, “Tuesday”, “Wednesday” or degrees: “high”, “medium”, “low”.

## Quantitative Data
There are divisions of quantitative data: interval and ratio.
- Interval data are continuous numerical values where the difference between values has meaning but there is no defined minimum and maximum value, like time, which has no beginning or end.
- Ratio data are continuous numerical values that have a defined zero point, like distances or temperature that begins at absolute zero.  

There are specific [statistics](https://www.statisticssolutions.com/free-resources/directory-of-statistical-analyses/data-levels-and-measurement/) relevant to each type.
- For categorical data, the mode and frequency distributions are relevant.
- For ordinal data in addition to those, the median or central value, minimum, maximum, range, and percentiles are relevant.
- Interval data points can be added and subtracted, and the difference between, mean, variance, and histograms of values are relevant.
- Ratio data points can also be multiplied and divided, and ratios, correlation and regression statistics are relevant.

## Booleans

Booleans are the result of evaluating conditions or comparisons. They have two values: True and False. They can be considered Nominal data or a label of two separate states. A number of operations return booleans, such as comparison operators equals '==', not equals '!=', and numeric comparison operators:
- \> (greater than)
- \>= (greater than or equal)
- < (less than)
- <= (less than or equal)

```python
print("A true condition has the value ", 4 > 3)
print("A false condition has the value ", "car" == "bus")
```

## Strings
Strings are sequences of text characters inside matching single or double quotes. Here are some example strings:

```python
print('a string inside single quotes')
print("a string inside double quotes")
print('single quotes used to include "double quotes" within a string' )
print("double quotes used to include 'single quotes' within a string")
```

The escape character \ changes the meaning of the following character so it is treated literally:

```python
print('one way to include both \'single\' and "double" quotes within a string')
print("another way to include both 'single' and \"double\" quotes within a string")
```

Triple double """ or single ''' quotes surrounding a string cause all characters between to be treated literally:

```python
print('''both 'single' and "double" quotes need not be escaped''')
print("""both 'single' and "double" quotes need not be escaped in this case either""")
```

# Python Collection Data Types
Python collection data types save groups of data as one value. These include:
- Lists: an ordered sequence of any values where the values can be added, changed, or deleted
- Tuples: like a list but cannot be changed
- Sets: a group of values with no duplicates
- Dictionaries: a group where values are associated with a key

Here are details of the types.


## Lists
Lists are dynamic series of values that can be accessed by position in the sequence. Lists are formed with enclosing values of any type between square brackets []. List values are accessed by position counting from 0:
- list[0] is the first element
- list[1] is the second element

and so on. There are a variety of operations that can be performed on lists.  

```python
# create a list
list = ["string", 1, 4.2]
print("1. ", list)
print("2. ", list[0])
print("3. ", list[2]*2)
# count the number of values in the list
print("4. ", len(list))
# change a value in a list
list[1] = 5
print("5. ", list)
# adding values to a list
list.append("new last value")
print("6. ", list)
# removing a value from a list
list.remove(4.2)
print("7. ", list)
# adding a value at a specific position
list.insert(0, 1)
print("8. ", list)
# count the number of a value in a list
print("9. ", list.count(1))
# finding the position of a value in the list
print("10. ", list.index("string"))
# reversing a list
list.reverse()
print("11. ", list)
# sorting a list
numbers = [3, 12, 4.5, 10]
numbers.sort()
print("12. ", numbers)
# detecting if a value is in a list
print("13. ", "string" in list)
print("14. ", 8 not in numbers)
# combining two lists
print("15. ", list + numbers)
# repeating a list
print("16. ", numbers * 2)
# numeric operations on list: min
print("17. ", min(numbers))
# numeric operations on list: max
print("18. ", max(numbers))
# numeric operations on list: sum
print("19. ", sum(numbers))
# remove all items in a list
numbers.clear()
print("20. ", numbers)
# extracting part of a list
print("21. ", list[1:3])
# printing the first elements of a list
print("22. ", list[:2])
# printing the last elements of a list
print("23. ", list[2:])
# printing every other element of a list
print("24. ", list[0:3:2])
# indexing from the back of a list
print("25. ", list[-1])
```

## Tuples
The purpose of tuples is to have a list that cannot be accidentally changed. This is partly a safety feature, many people may work on a Python program and there is value in preventing mistakes.

Tuple are formed with enclosing values of any type between parentheses (). Tuple values are also accessed by position counting from 0. Tuple operations are the same as for lists but not including the ones that alter the tuple in any way.  

```python
# create a tuple
tuple = ("string", 1, 4.2)
print("1. ", tuple)
print("2. ", tuple[0])
print("3. ", tuple[2]*2)
# count the number of values in the tuple
print("4. ", len(tuple))
# count the number of a value in a tuple
print("5. ", tuple.count(1))
# finding the position of a value in the tuple
print("6. ", tuple.index("string"))
# detecting if a value is in a tuple
numbers = (3, 12, 4.5, 10)
print("7. ", "string" in tuple)
print("8. ", 8 not in numbers)
# combining two tuples
print("9. ", tuple + numbers)
# repeating a tuple
print("10. ", numbers * 2)
# numeric operations on tuple: min
print("11. ", min(numbers))
# numeric operations on tuple: max
print("12. ", max(numbers))
# numeric operations on tuple: sum
print("13. ", sum(numbers))
# extracting part of a tuple
print("14. ", tuple[1:3])
# printing the first elements of a tuple
print("15. ", tuple[:2])
# printing the last elements of a tuple
print("16. ", tuple[2:])
# printing every other element of a tuple
print("17. ", tuple[0:3:2])
# indexing from the back of a tuple
print("18. ", tuple[-1])
```

## Sets
Sets are an unordered group of values with no duplicates. Sets have rich set of operations as defined by [set theory](https://realpython.com/python-sets/), of which a useful set are implemented in Python.

```python
basket1 = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
# show that duplicates have been removed
print("1.", basket1)
# count the number of values in the set
print("2.", len(basket1))
# detecting if a value is in a set
print("3.", 'orange' in basket1)
print("4.", 'crabgrass' in basket1)
basket2 = {'pear', 'orange', 'lemon', 'banana', 'orange'}
# basket2 without duplicates
print("5.", basket2)
# set intersection: fruit in both basket1 and basket2
print("6.", basket1 & basket2)
# set union: fruit in basket1 or basket2 or both
print("7.", basket1 | basket2)
# set difference: fruit in basket1 but not in basket2
print("8.", basket1 - basket2)
# fruit in basket1 or basket2 but not both
print("9.", basket1 ^ basket2)
basket3 = {'pear', 'lemon', 'banana'}
# basket3 is a subset of basket2
print("10.", basket3 <= basket2)
# basket2 is a superset of basket3
print("11.", basket2 >= basket3)
# adding a new element to set
basket1.add('peach')
print("12.", basket1)
# adding an existing element to set
basket1.add('orange')
print("13.", basket1)
# removing an element from a set
basket1.remove('peach')
print("14.", basket1)
# removing all elements from a set
basket1.clear()
print("15.", basket1)
```

## Dictionaries
Dictionaries are a group where you look up a value with another value, a key, like looking up a word definition in a real dictionary. Dictionaries are formed with enclosing pairs of values of any type between braces {}. Dictionary values are accessed by key using square brackets.


```python
# create a dictionary
dictionary1 = {"leaf": "green", "sky": "blue", "apple": "red"}
print("1. ", dictionary1)
print("2. ", dictionary1["leaf"])
# count the pairs in a dictionary
print("3. ", len(dictionary1))
# change a values in a dictionary
dictionary1["apple"] = "yellow"
print("4. ", dictionary1)
# adding values to a dictionary
dictionary1["earth"] = "brown"
print("5. ", dictionary1)
# removing a value from a dictionary
del dictionary1["sky"]
print("6. ", dictionary1)
#  detecting if a key is in a dictionary
print("7. ", "leaf" in dictionary1)
print("8. ", "leaf" not in dictionary1)
# list all the keys in a dictionary
print("9. ", dictionary1.keys())
# list all the values in a dictionary
print("10. ", dictionary1.values())
# combine two dictionaries
dictionary2 = {"apple": "yellow", "sunset": "orange"}
dictionary1.update(dictionary2)
print("11. ", dictionary1)
# remove all pairs in a dictionarie
dictionary1.clear()
print("12. ", dictionary1)
```
