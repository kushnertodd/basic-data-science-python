# Basic Python Data Types
Python has four basic built-in types of data:
- Booleans: validity that condition holds
- Strings: series of text characters
- Integers: negative and positive whole numbers, without fractions
- Floats: continuous numbers with fractions

## Booleans

Booleans are the result of evaluating conditions or comparisons. They have two values: True and False. They can be considered Nominal data or a label of two separate states. A number of operations return booleans, such as comparison operators equals '==', not equals '!=', and numeric comparison operators:
- \> (greater than)
- \>= (greater than or equal)
- < (less than)
- <= (less than or equal)

```python
print("A true condition has the value ", 4 > 3)
print("A false condition has the value ", "car" == "bus")
```
